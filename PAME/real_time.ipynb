```python
import requests
import sqlite3
import time

# ThingSpeak API Details
channel_id = '2492518'
read_key = 'ELOPSF1X9GS7O0WN'

def fetch_data_from_thingspeak():
    """Fetches data from ThingSpeak channel."""
    url = f'https://api.thingspeak.com/channels/{channel_id}/feeds.json?api_key={read_key}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data['feeds']  
    return []

import time

def store_data_to_sqlite(data):
    """Stores data into SQLite database."""
    db_file = 'places.db'
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()
    retry_count = 0
    max_retries = 3
    
    while retry_count < max_retries:
        try:
            cursor.execute('''CREATE TABLE IF NOT EXISTS places
                              (id INTEGER PRIMARY KEY AUTOINCREMENT, 
                               name REAL, details REAL, image REAL)''')
            
            # Insert data
            for entry in data:
                loc = entry.get('field1')
                dis = int(entry.get('field2') or 0) / 100
                cam_dis = int(entry.get('field3') or 0) / 100  # Using 'or 0' to handle None
                area = int(entry.get('field4') or 0) / 10000
                volume = area * (dis - cam_dis + 2)
                cost = volume * 3400
                text = f'Location:{loc},area:{area}sqcm,volume:{volume}cubic meter,cost:{cost}Rs'
                cursor.execute('INSERT INTO places (name, details, image) VALUES (?, ?, ?)',
                               ('place', text, 'img_loc'))
            
            conn.commit()
            break  # Operation successful, exit loop
        except sqlite3.OperationalError as e:
            if "database is locked" in str(e):
                # Database is locked, retry after a short delay
                retry_count += 1
                time.sleep(1)  # Wait for 1 second before retrying
            else:
                # Unexpected error, raise it
                raise e
        finally:
            cursor.close()
            conn.close()

    if retry_count == max_retries:
        print("Maximum retries exceeded. Unable to write to the database.")

    
def main():
    while True:  # Loops to fetch and store data periodically
        data = fetch_data_from_thingspeak()
        if data:
            store_data_to_sqlite(data)
        time.sleep(60*15)  # Pause for 15 minutes before fetching again

if __name__ == "__main__":
    main()
```
